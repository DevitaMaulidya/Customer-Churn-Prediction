```python
import pandas as pd
import numpy as np
import random
from faker import Faker

fake = Faker()
np.random.seed(42)
random.seed(42)
n = 1000
max_attempts = 5000

seen_surnames = set()
unique_surnames = []

# Hindari duplikat nama
attempts = 0
while len(unique_surnames) < n and attempts < max_attempts:
    name = fake.last_name()
    if name not in seen_surnames:
        seen_surnames.add(name)
        unique_surnames.append(name)
    attempts += 1
    
# Buat data
customer_ids = list(range(15600000, 15600000 + n))
surnames = list({fake.unique.last_name() for _ in range(n * 2)})[:n]
credit_scores = np.clip(np.random.normal(650, 50, n).astype(int), 300, 850)
geography = np.random.choice(['France', 'Spain', 'Germany'], size=n, p=[0.4, 0.3, 0.3])
gender = np.random.choice(['Male', 'Female'], size=n)
ages = np.clip(np.random.normal(40, 10, n).astype(int), 18, 92)
tenure = np.random.randint(0, 11, size=n)
balances = np.round(np.where(ages > 40, np.random.normal(100000, 50000, n), np.random.normal(50000, 30000, n)), 2)
balances = np.clip(balances, 0, 250000)
num_products = np.random.choice([1, 2, 3, 4], size=n, p=[0.5, 0.3, 0.15, 0.05])
has_credit_card = np.random.choice([0, 1], size=n, p=[0.3, 0.7])
is_active = np.random.choice([0, 1], size=n, p=[0.5, 0.5])
salaries = np.round(np.random.uniform(10000, 150000, size=n), 2)
complain = np.random.choice([0, 1], size=n, p=[0.8, 0.2])
satisfaction = np.random.choice([1, 2, 3, 4, 5], size=n, p=[0.1, 0.2, 0.4, 0.2, 0.1])
card_type = np.random.choice(['SILVER', 'GOLD', 'PLATINUM', 'DIAMOND'], size=n, p=[0.3, 0.3, 0.25, 0.15])
points = (satisfaction * 100 + balances / 1000).astype(int)

# Gabung ke DataFrame
df = pd.DataFrame({
    'CustomerId': customer_ids,
    'Surname': surnames,
    'CreditScore': credit_scores,
    'Geography': geography,
    'Gender': gender,
    'Age': ages,
    'Tenure': tenure,
    'Balance': balances,
    'NumOfProducts': num_products,
    'HasCrCard': has_credit_card,
    'IsActiveMember': is_active,
    'EstimatedSalary': salaries,
    'Complain': complain,
    'SatisfactionScore': satisfaction,
    'CardType': card_type,
    'PointEarned': points
})

df.to_csv("synthetic_customer_data_clean.csv", index=False)
```
